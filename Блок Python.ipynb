```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Прочитайте данные (переменную назовите 'df')
df = pd.read_csv('data.csv')

# Вывести несколько первых строк таблицы данных
print(df.head())
```

Проверяем формат столбцов

```python
print(df.info())
```

Сразу переведем столбец "Дата" в правильный формат

```python
df['Дата'] = pd.to_datetime(df['Дата'])
```

```python
df
```

Сгруппируйте данные по дате, посчитайте количество продаж

```python
grouped_df = df.groupby('Дата')['Количество'].sum()
grouped_df
```

Вывести несколько первых строк сгруппированных данных

```python
grouped_df.head()
```

Нарисуйте график продаж у `grouped_df`

```python
plt.figure(figsize=(10, 6))  
plt.plot(grouped_df, marker='o', linestyle='-') 
plt.title('Продажи по датам')  
plt.xlabel('Дата')  
plt.ylabel('Количество продаж') 
plt.xticks(rotation=45)  
plt.grid(True) 
plt.tight_layout() 
plt.show()  
```

Опишите что вы видите на графике. Ваша задача - максимально описать график

Продажи колеблются в течение рассматриваемого периода, что можно увидеть по вертикальным линиям, соединяющим точки данных. Пики и впадины указывают на изменчивость объемов продаж от даты к дате.
Объемы продаж колеблются примерно от 3000 до более чем 6000 единиц, что говорит о значительных вариациях в объемах продаж в течение рассматриваемого периода.
По графику сложно определить явную тенденцию (тренд) увеличения или уменьшения объемов продаж в течение года, также не видна явная сезонность. 
На графике присутствуют точки, которые выделяются за пределы общего диапазона колебаний, что может указывать на аномально высокие или низкие значения продаж в определенные даты.
Точки данных похоже распределены равномерно по времени, что может означать, что данные о продажах собирались регулярно (ежедневно).

Найдите строку, у которой максимальный выброс по количеству продаж (нужно найти выброс у `df`)

```python
max_sales_row = df[df['Количество'] == df['Количество'].max()]
print(max_sales_row)
```

```python
Q1 = df['Количество'].quantile(0.25)
Q3 = df['Количество'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Количество'] < lower_bound) | (df['Количество'] > upper_bound)]
print(outliers)
```

```python
higher_outliers = df[(df['Количество'] > upper_bound)]
lower_outliers = df[df['Количество'] < lower_bound]
max_outlier = higher_outliers[higher_outliers['Количество'] == higher_outliers['Количество'].max()]
print(max_outlier)
if not lower_outliers.empty:
    min_outlier = lower_outliers[lower_outliers['Количество'] == lower_outliers['Количество'].min()]
    print(f"Самый низкий выброс: {min_outlier}")
else:
    print("Выбросы ниже нижней границы отсутствуют.")
```

Найдите топовый товар по продажам по средам за июнь, июль, август у 3 склада

```python
df = df[df['Склад'] == 3]
df = df[df['Дата'].dt.dayofweek == 2]
df = df[df['Дата'].dt.month.isin([6, 7, 8])]
grouped = df.groupby('Номенклатура')['Количество'].sum()
top_product = grouped.idxmax()
print(f"Топовый товар по продажам по средам за июнь, июль, август у 3 склада: {top_product}")
```
